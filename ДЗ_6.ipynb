```python
import pandas as pd
import numpy as np



url = 'https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv'
alcohol_df = pd.read_csv(url)
```

```python
alcohol_df.columns
```

```python
#Find the number of countries where the beer servings equals to 0?

alcohol_df[(alcohol_df['beer_servings'] == 0)]
```

```python
#Find the number of countries where all kind of alcohol consumption is equal to 0?

alcohol_df[(alcohol_df['total_litres_of_pure_alcohol'] == 0)]
```

```python
#Find the number of countries where the total litres of pure alcohol is higher or equal to 9?

alcohol_df[(alcohol_df['total_litres_of_pure_alcohol'] >= 9)]
```

```python
#Find the country where people drink the most beer, wine or spirits compared to other countries?

max_beer = alcohol_df['beer_servings'].max()
max_wine = alcohol_df['wine_servings'].max()
max_spirit = alcohol_df['spirit_servings'].max()

print (alcohol_df[(alcohol_df['beer_servings'] == max_beer)])
print (alcohol_df[(alcohol_df['wine_servings'] == max_wine)])
print (alcohol_df[(alcohol_df['spirit_servings'] == max_spirit)])
```

```python
# max_beer - Namibia, max_wine - France, max_spirit - Grenada
```

```python
#Compute the average of pure alcohol consumption in countries which end with a suffix -stan

pure_alc = alcohol_df[alcohol_df['country'].str.endswith("stan")]
print("average of pure alcohol" , pure_alc['total_litres_of_pure_alcohol'].mean())
```
