```python
import pandas as pd
import numpy as np



call_center_data = pd.read_csv('/Call Center Data.csv')
abandoned_calls = call_center_data['Incoming Calls'] - call_center_data['Answered Calls']
call_center_data ['Abandoned Calls'] = abandoned_calls
call_center_data ['Efficiency of Employees (%)'] = round (call_center_data['Answered Calls'] / call_center_data['Incoming Calls'], 2)
call_center_data ['Rate (%) of  Abandonment of Calls'] = round (call_center_data ['Abandoned Calls'] / call_center_data['Incoming Calls'], 2)
call_center_data.tail()


```

```python
fast_employees = call_center_data[(call_center_data['Waiting Time (AVG)'] < '00:10:00')]
fast_employees.tail()
```

```python
medium_employees = call_center_data[(call_center_data['Waiting Time (AVG)'] > '00:10:00') & (call_center_data['Waiting Time (AVG)'] <= '00:20:00') ]
medium_employees.tail()
```

```python

slow_employees = call_center_data[(call_center_data['Waiting Time (AVG)'] > '00:20:00')]
slow_employees
```

```python
#How many employees have a Telephone Answering Rate of more than 80%
call_center_data = call_center_data[(call_center_data['Efficiency of Employees (%)'] > 0.80)]
print('employees have a Telephone Answering Rate of more than 80% is:', len(call_center_data.index))
```

```python
#Based on the new employee call answering speed category; what percentage of employees fall in the category of MEDIUM?
print ('Medium employess percentage is:', round(len(medium_employees.index)/len(call_center_data.index)*100, 4), '%')
```

```python
#How many employees receive less than 50 calls?
call_center_data = call_center_data[(call_center_data['Incoming Calls'] < 50)]
print ('employees receive less than 50 calls is:', len(call_center_data.index))
```

```python
#What percentage of employees who have a Talk Duration of more than 2 minutes (120 seconds)?
talk_duration_120sec = call_center_data[(call_center_data['Talk Duration (AVG)'] > '00:02:00')]
print('percentage of employees who have a Talk Duration of more than 2 minutes is:', round (len(talk_duration_120sec.index) /len(call_center_data.index)*100,4), '%')
```
